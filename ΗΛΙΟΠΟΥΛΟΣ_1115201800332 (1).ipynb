# [ 3 ] Gradient Decent


## [ 3.1 ]

```python
def f(x1,x2):
  return pow((x1-2),2)+pow((x2-3),2)

def df1(x):
  return 2*(x-2)

def df2(x):
  return 2*(x-3)

def GradientDecent(lr,precision,startX1,startX2,maxIters):
  learning_rate = lr
  precision = precision
  curX1, curX2 = startX2, startX2
  maxIters = maxIters

  values = []
  iters = []

  for i in range(maxIters):
    prevX1 = curX1
    prevX2 = curX2

    values.append(f(curX1,curX2))
    print("iteration=",i,"f=",f(curX1,curX2))

    curX1 = curX1 - learning_rate * df1(prevX1)
    curX2 = curX2 - learning_rate * df2(prevX2)

    diff = abs(curX1-prevX1) + abs(curX2-prevX2)

    iters.append(i)
    if diff <= precision:
      print("iteration=",i,"The local minimum occurs at curX1=", curX1, "curX2=",curX2,"f=",f(curX1,curX2))
      break

  plt.title("Value of f(x1,x2) in every iteration")
  plt.plot(iters, values, label="f(x1,x2)")
  plt.xlabel("Iterations")
  plt.ylabel("Value")
  plt.legend(loc='best')
  plt.show()
```

Parameters:

- Precision: 0.0001
- curX2: 0
- curX2: 0
- maxIters: 10

Βρήκε την απάντηση κατευτείαν

## [ 3.2 ]

```python
def f(x1,x2):
  return pow((1-(x2-3)),2) + 20*pow(((x1+3)-pow((x2-3),2)),2)

def df1(x1,x2):
  return 40*(x1-((x2-3)*(x2-3))+3)

def df2(x1,x2):
  return 2*((-40)*(x2-3)*(x1-((x2-3)*(x2-3))+3) + x2-4)

def GradientDecent(lr,precision,startX1,startX2,maxIters):
  learning_rate = lr      # Learning rate
  precision = precision   # This tells us when to stop the algorithm
  curX1 = startX1         # The algorithm starts at x1=0
  curX2 = startX2         # The algorithm starts at x2=0
  maxIters = maxIters     # maximum number of iterations

  values = []
  iters = []

  for i in range(maxIters):
    prevX1 = curX1
    prevX2 = curX2

    values.append(f(curX1,curX2))
    print("iteration=",i,"f=",f(curX1,curX2))
    curX1 = curX1 - learning_rate * df1(prevX1,prevX2)
    curX2 = curX2 - learning_rate * df2(prevX1,prevX2)

    diff = abs(curX1-prevX1) + abs(curX2-prevX2)

    iters.append(i)
    if diff <= precision:
      print("iteration=",i,"The local minimum occurs at curX1=", curX1, "curX2=",curX2,"f=",f(curX1,curX2))
      break

  plt.title("Value of f(x1,x2) in every iteration")
  plt.plot(iters, values, label="f(x1,x2)")
  plt.xlabel("Iterations")
  plt.ylabel("Value")
  plt.legend(loc='best')
  plt.show()
```
